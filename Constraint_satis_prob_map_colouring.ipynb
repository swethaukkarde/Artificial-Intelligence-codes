```
# List of regions
regions = ['WA', 'NT', 'SA', 'Q', 'NSW', 'V', 'T']

# Available colours
colours = ['Red', 'Green', 'Blue']

# Adjacency constraints
adjacent = {
    'WA': ['NT', 'SA'],
    'NT': ['WA', 'SA', 'Q'],
    'SA': ['WA', 'NT', 'Q', 'NSW', 'V'],
    'Q': ['NT', 'SA', 'NSW'],
    'NSW': ['Q', 'SA', 'V'],
    'V': ['SA', 'NSW'],
    'T': []
}

# Check if assigning colour is valid
def is_safe(region, colour, assignment):
    for neighbour in adjacent[region]:
        if neighbour in assignment and assignment[neighbour] == colour:
            return False
    return True

# Backtracking function
def map_colouring(assignment):
    # If all regions are coloured
    if len(assignment) == len(regions):
        return assignment

    # Select an unassigned region
    for region in regions:
        if region not in assignment:
            break

    # Try all colours
    for colour in colours:
        if is_safe(region, colour, assignment):
            assignment[region] = colour

            result = map_colouring(assignment)
            if result:
                return result

            # Backtrack
            del assignment[region]

    return None

# Solve the problem
solution = map_colouring({})

print("Colour Assignment Solution:")
for region, colour in solution.items():
    print(region, "->", colour)
```
