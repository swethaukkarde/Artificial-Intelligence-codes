```
import math

def alpha_beta(depth, node_index, is_maximizing, values, alpha, beta):
    # Terminal condition
    if depth == 0:
        return values[node_index]

    if is_maximizing:
        best = -math.inf

        for i in range(2):
            val = alpha_beta(
                depth - 1,
                node_index * 2 + i,
                False,
                values,
                alpha,
                beta
            )
            best = max(best, val)
            alpha = max(alpha, best)

            # Pruning
            if beta <= alpha:
                break

        return best

    else:
        best = math.inf

        for i in range(2):
            val = alpha_beta(
                depth - 1,
                node_index * 2 + i,
                True,
                values,
                alpha,
                beta
            )
            best = min(best, val)
            beta = min(beta, best)

            # Pruning
            if beta <= alpha:
                break

        return best


# Driver code
values = [3, 5, 6, 9, 1, 2, 0, -1]
depth = 3

result = alpha_beta(
    depth,
    0,
    True,
    values,
    -math.inf,
    math.inf
)

print("Optimal value:", result)
```
