```
def backward_chaining(goal, facts, rules, visited=None):
    if visited is None:
        visited = set()

    # If goal is already a known fact
    if goal in facts:
        return True

    # Prevent infinite loops
    if goal in visited:
        return False

    visited.add(goal)

    # Try to satisfy goal using rules
    for premises, conclusion in rules:
        if conclusion == goal:
            # Check if all premises can be proven
            if all(backward_chaining(p, facts, rules, visited) for p in premises):
                return True

    return False
```

```
# Known facts
facts = {'A', 'B'}

# Rules: (premises â†’ conclusion)
rules = [
    (['A', 'B'], 'C'),
    (['C'], 'D'),
    (['D'], 'E')
]

# Goal to prove
goal = 'E'

# Run backward chaining
if backward_chaining(goal, facts, rules):
    print("Goal", goal, "is PROVED")
else:
    print("Goal", goal, "cannot be proved")
```
